```python
input_path = "/Volumes/ppe/ppe/ppe_input"
archive_path = "/Volumes/ppe/ppe/ppe_archived"
```

```python
files = dbutils.fs.ls(input_path)

if not files:
    print("No files found in input folder. Nothing to archive.")
    dbutils.notebook.exit("NO_FILES")
```

```python
dbutils.fs.mkdirs(archive_path)
```

```python
for f in files:
    if f.isFile():
        source_file = f.path
        target_file = f"{archive_path}/{f.name}"

        print(f"Archiving file: {f.name}")
        dbutils.fs.mv(source_file, target_file)
```

```python
print("All source files successfully archived.")
```
